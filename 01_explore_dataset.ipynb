### Identify subcategories

```python
types_df={}
```

```python
for long_value in types:
    for short_key in types_smallest:
        if short_key in long_value:
            types_df[long_value] = short_key
```

```python
categories=pd.DataFrame.from_dict(types_df.items())
```

```python
categories.columns=['TITLE','REF_AREA']
```

```python
categories
```

```python
categories['granuality'] = categories.TITLE.apply(lambda x: x.split(' - ')[1].split('based on')[0].strip(' ') if (' - ' in x and 'based on' in x ) else pd.NA)
```

```python
categories.granuality.value_counts()
```
