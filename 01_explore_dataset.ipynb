```python
eur.TITLE.value_counts()
```

```python
len(eur[eur.REF_AREA=='HU'].TITLE.unique()) # HU has all value, yey!
```

```python
types=sorted(eur.TITLE.unique())
```

```python
types_unique={i.split('-')[0].strip(' ') for i in types}
```

```python
types_smallest={i.split(' - ')[0].strip(' ') for i in types}
types_smallest=list(types_smallest)
types_smallest={i.split(', ')[0].strip(' ') for i in types_smallest}
```

```python
types_smallest
```

```python
len(types_smallest)
```

We have 20 unique indicators, and these are splitted to 563 subcategories.

### Identify subcategories

```python
types_df={}
```

```python
for long_value in types:
    for short_key in types_smallest:
        if short_key in long_value:
            types_df[long_value] = short_key
```

```python
categories=pd.DataFrame.from_dict(types_df.items())
```

```python
categories.columns=['TITLE','REF_AREA']
```

```python
categories
```

## Time period for each countries

```python
countries=pd.pivot_table(eur[['REF_AREA','TIME_PERIOD', 'TITLE']], index='REF_AREA', columns='TIME_PERIOD', values='TITLE',aggfunc='count')
```

```python
countries=countries.notna()
```

```python
sns.heatmap(data=countries, cmap='Greys')
plt.axvline(x=22, color='red', linewidth=2)
plt.show()
```

```python
countries.columns=='2014-Q3'
```

```python
countries_df=pd.melt(countries, id_vars='REF_AREA', var_name='TIME_PERIOD', value_name='COUNTRY')
```

```python
countries_df['COUNTRY']=countries_df.COUNTRY.astype(int)
```
